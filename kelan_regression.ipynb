```python
import pandas as pd
import numpy as np
import stat_utils
import statsmodels.formula.api as smf
import statsmodels.api as sm
import re
import scipy
import patsy
from pprint import pprint
```

##Load clean, and filter the data

```python
raw_data = pd.read_csv(r"Portland Regression FINAL.csv")
```

```python
raw_data.columns = [x.strip() for x in raw_data.columns.tolist()] #Remove the white space from arround column names
```

```python
print "The number of non_null rent observations is {}".format(len(raw_data[pd.notnull(raw_data.rent)]))
print "The total number of observations is {}".format(len(raw_data))
```

```python
raw_data.columns #Display at the order of the columns
```

```python
#Iterate over each observation and each column from bedrooms to the end, and for each value that is a string
#remove any $ signs and any commas, and try to convert the value to a number
for ind, row in raw_data.copy().iterrows(): 
    for col in raw_data.columns[5:]:
        item_val = row[col]
        if isinstance(item_val, str):
            converted_val = re.sub("[$,]", "", item_val)
            converted_val = converted_val.strip()
            try:
                raw_data.loc[ind, col] = float(converted_val)
            except:
                pass
            
#For the bedroom column and all columns afterwards, check if all values in that columns are numbers or are null values
#If they are, convert that column type to be float (aka, a column of numbers)
for col in raw_data.columns[5:]:
    cond1 = np.array([x*0 == 0 for x in raw_data[col]]) #For each entry, check if it is a number
    cond2 = np.array([pd.isnull(x) for x in raw_data[col]]) #For each entry, check if it is "Not A Number", i.e is null.
    if all(cond1 + cond2 - cond1 * cond2): #Check if all entries are a number or are null
        raw_data[col] = raw_data[col].astype(float) #If so, convert the entire column to a numeric type
```

```python
#Convert the dash values ("-") in the parking per unit column to null values and change the column tpe to float
for ind, x in raw_data.parkingperunit.copy().iteritems():
    if isinstance(x, str):
        conv_str = x.strip()
        if "-" in conv_str:
            new_val = np.nan
        else:
            new_val = float(conv_str)
    else:
        new_val = x
    raw_data.parkingperunit.at[ind] = new_val
    
raw_data.parkingperunit = raw_data.parkingperunit.astype(float)
```

```python
#Filter out data with null rent values and with null parking per unit values
non_null_data = raw_data[pd.notnull(raw_data.rent) & pd.notnull(raw_data.parkingperunit)]
```

####Define helper functions for model diagnostics

```python
#Create a function to perform different statistical tests of normality for model residuals
def many_normality_tests(resids):
    print "The schapiro-wilks test statistic and its p-value:"
    pprint(scipy.stats.shapiro(resids))
    print "="*10
    print "The anderson-darling test statistic, and associated critical values for the p-values (last two rows):"
    pprint(scipy.stats.anderson(resids))
    print "="*10
    print "The kolmogorov-smirnov test statistic and its p-value:"
    pprint(scipy.stats.kstest(resids, 'norm'))
```

```python
#Create a function to perform different statistical tests of heteroskedasticity for model residuals
def many_variance_tests(resids, explanatory_var_matrix, exog_formula = False):
    sep = "="*20
    if exog_formula:
        exog_matrix = patsy.dmatrix(exog_formula, explanatory_var_matrix)
    else:
        eformula = " + ".join(explanatory_var_matrix.columns.tolist())
        exog_matrix = patsy.dmatrix(eformula, explanatory_var_matrix)
    print "The results of the breush-pagan test for heteroskedasticity (alternating test_value, test_statistic) are:"
    pprint(sm.stats.diagnostic.het_breushpagan(resids, exog_matrix))
    print sep
    try:
        print "The results of the white test for heteroskedasticity (alternating test_value, test_statistic) are:"
        pprint(sm.stats.diagnostic.het_white(resids, exog_matrix))
    except:
        print "There was an error while running the white test for heteroskedasticity"
```

#Models and Results

###Model 1: (from your draft paper)

```python
m1_rhs = "sqft + units + transitscore + eastwest + parkingperunit"
mod1 = smf.ols("rent ~ " + m1_rhs, data = non_null_data).fit()
```

```python
mod1.summary()
```

<font color="red">See residual plots below, there is likely non-normality and heteroskedasticity</font>

```python
stat_utils.all_residual_plots(mod1, non_null_data)
```

####Test for normality and heteroskedasticity of the original model in Kelan's draft TRB paper.

According to the shapiro-wilk, the Anderson-Darling, and the Kolmogoroz-Smirnov tests, we should reject the null hypothesis that the residuals are normally distributed.

```python
many_normality_tests(mod1.resid)
```

Additionally, there is heteroskedasticity in the model (note the p-values of less than 0.05):

```python
many_variance_tests(mod1.resid, non_null_data, exog_formula=m1_rhs)
```
